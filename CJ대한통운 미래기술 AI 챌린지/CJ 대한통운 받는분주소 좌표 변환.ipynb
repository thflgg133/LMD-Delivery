```python
import requests; from urllib.parse import urlparse
import pandas as pd
import geopandas
 
list = [] 
for i in range(2836):
  sm_name = delivery_df['SM명'][i]
  address = delivery_df['받는분주소'][i]
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
  rest_api_key = '2b11620f039b990d063a49e80b9b6861'
  result = requests.get(urlparse(url).geturl(),
                        headers = {'Authorization': 'KakaoAK ' + rest_api_key})
  
  json_obj = result.json()

  if json_obj['documents'] == []:
    json_obj['documents'] = '0'

  for document in json_obj['documents']:
 
    if document == '0':
      val =[None, None, None, None]
      list.append(val)
      
      
    else:
      val = [sm_name, document['address_name'], document['y'], document['x']]
      list.append(val)
  
print(list)
```

```python
df = pd.DataFrame(list, columns=['SM명','받는분주소', 'lat', 'lon'])
df
df.to_csv('./drive/MyDrive/CJ대한통운 미래기술 AI 챌린지/택배 배송 데이터 CSV/좌표화 까지 완료한 CSV 데이터/2-1. 택배 배송 데이터.csv', index=False) # index=False를 해야 Unnamed: 0 안 붙음
```
