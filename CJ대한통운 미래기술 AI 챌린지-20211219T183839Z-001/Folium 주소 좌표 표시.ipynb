```python
import folium as g
import pandas as pd
```

```python
g_map = g.Map(location=[37.1439207424255,126.817480969444],zoom_start = 15)
g_map
```

```python
# 주소 할당

base_src = './drive/MyDrive/CJ대한통운 미래기술 AI 챌린지/택배 배송 데이터 CSV/좌표화 까지 완료한 CSV 데이터/'
delivery_data = base_src + '2-1. 택배 배송 데이터.csv'
delivery_df = pd.read_csv(delivery_data, encoding='utf-8')
delivery_df
```

```python
delivery_df.dropna(axis=0,inplace=True)
```

```python
delivery_df.reset_index(inplace=True)
```

```python
delivery_df
```

```python
delivery_map = g.Map(location=[37.1439207424255,126.817480969444],zoom_start=15)
delivery_map
```

```python
for i in range(100):
    marker = g.Marker([delivery_df.loc[i]['lat'], delivery_df.loc[i]['lon']],
                    icon = g.Icon(color='blue'))

    marker.add_to(delivery_map)
```

```python
delivery_map
```

```python
delivery_map.save("delivery_map.html")
```
